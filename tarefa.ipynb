```python
# Importação das bibliotecas necessárias
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

# Exibir gráficos inline
%matplotlib inline
```

```python
# Carregar imagem de exemplo
from PIL import Image
import requests

url = "https://images.unsplash.com/photo-1745810187217-4d9e1ccfd9d5?ixlib=rb-4.1.0&q=85&fm=jpg&crop=entropy&cs=srgb&w=640"
image = Image.open(requests.get(url, stream=True).raw)
image = image.convert("RGB")  # Garante 3 canais

# Mostrar imagem
plt.imshow(image)
plt.title("Imagem Original")
plt.axis('off')
plt.show()
```

```python
np_image = np.array(image)
print("Formato da imagem:", np_image.shape)
print("Valor do pixel (0, 0):", np_image[0, 0])
```

```python
red_channel = image.getchannel('R')

plt.imshow(red_channel, cmap='Reds')
plt.title("Canal Vermelho")
plt.axis('off')
plt.show()
```

```python
image_array = np.array(image)

image_array[:, :, 0] = np.random.randint(0, 256, size=image_array[:, :, 0].shape, dtype=np.uint8)

modified_red_image = Image.fromarray(image_array)

plt.imshow(modified_red_image)
plt.title("Canal Vermelho Aleatório")
plt.axis('off')
plt.show()
```

```python
negative_image = Image.fromarray(255 - np.array(image))

plt.imshow(negative_image)
plt.title("Imagem Negativa")
plt.axis('off')
plt.show()
```

```python
negative_image.save("imagem_negativa.jpg")
print("Imagem negativa salva como 'imagem_negativa.jpg'")
```

```python
crop_box = (0, 0, 100, 100)
cropped_image = image.crop(crop_box)

cropped_image.save("imagem_recortada.jpg")
print("Imagem recortada salva como 'imagem_recortada.jpg'")
```
